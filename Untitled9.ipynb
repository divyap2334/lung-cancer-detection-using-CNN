```python

import numpy as np
import pydicom
#from pydicom.data import get_testdata_file
import pylab
import os
import matplotlib.pyplot as plt
import cv2
import math
#import pypng
%matplotlib inline
```

```python
PNG=False
dcm_path="D:/malignant_n"
png_path="D:/main_data/malignant_png"


dcm_path1="D:/non_malignant_n"
png_path1="D:/main_data/non_malignant_png"

```

```python
test_list1 = [ f for f in  os.listdir(dcm_path)]

print(test_list1)

for f in test_list1: 
    try:
        ds = pydicom.read_file(os.path.join(dcm_path , f),force=True) # read dicom image
        ds.file_meta.TransferSyntaxUID = pydicom.uid.ImplicitVRLittleEndian
        img = ds.pixel_array # get image array
        plt.imshow(img)
        cv2.imwrite(os.path.join(png_path , f.replace('.dcm','.png')), img)
        #cv2.imwrite(png_path + f.replace('.dcm','.png'),img) # write png image
    except Exception as e:
        pass
```

```python
test_list = [ f for f in  os.listdir(dcm_path1)]

print(test_list)

for f in test_list:  
    try:
        ds = pydicom.read_file(os.path.join(dcm_path1 , f)) # read dicom image
        ds.file_meta.TransferSyntaxUID = pydicom.uid.ImplicitVRLittleEndian
        img = ds.pixel_array # get image array
        plt.imshow(img)
        cv2.imwrite(os.path.join(png_path1 , f.replace('.dcm','.png')), img)
        #cv2.imwrite(png_path + f.replace('.dcm','.png'),img) # write png image
    except Exception as e:
        pass
```
